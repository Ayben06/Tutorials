```python
# gəlin bu əməliyyatları bütün sütuna tətbiq edək
# map funskiyasından istifadə edərək, stringi dictionary-ə çeviririk
df.loc['resistance'] = df.loc['resistance'].map(eval)
```

```python
# lakin nəyəsə görə bizdə xəta çıxır
```

```python
# bizi marağlandıran sütunda 8 NAN qiyməti var.
df['resistance'].isna().sum()
```

```python
# biz boolean mask istifadə edərək, NAN olmayan qiymətlərinə eval funksiyasını tətbiq edirik.
non_na_mask = ~df['resistance'].isna()
df.loc[non_na_mask,'resistance'] = df.loc[non_na_mask,'resistance'].map(eval)
```

```python
# sütunu təmizləmək üçün funskiya yazırıq
def clean_column(row, colname):
    
    val_dict = dict()
    for key, value in row.items():
        for cat in value:
            val_dict[f'{cat} {colname}'] = float(key.replace('%',''))
    return val_dict
```

```python
# təmiz data
df.loc[non_na_mask,'resistance'].apply(clean_column, colname='resistance')
```

```python
# dataframe şəklində təmiz data
df.loc[non_na_mask,'resistance'].apply(clean_column, colname='resistance').apply(pd.Series)
```

```python
# biz təzə datanı bir dataframe-də saxlayırıq.
new_df = df.loc[~df['resistance'].isna(),'resistance'].apply(clean_column, colname='resistance').apply(pd.Series)
```

```python
# və əsas datamızla JOIN edirik.
df = pd.merge(df, new_df, left_index=True, right_index=True)
```

```python
# weakness sütunu üçün
df.loc[~df['weakness'].isna(),'weakness'] = df.loc[~df['weakness'].isna(),'weakness'].map(eval)
new_df2 = df.loc[~df['weakness'].isna(),'weakness'].apply(clean_column, colname='weakness').apply(pd.Series)
df = pd.merge(df, new_df2, left_index=True, right_index=True)
```
