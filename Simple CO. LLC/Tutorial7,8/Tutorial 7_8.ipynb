Gördüyümüz kimi, yuxarıdakı kodla eyni nəticə alırıq. Then why to use "else"? \
Python dokumentasiyasında yazılıb ki, yaxşı olar ki, xətaya səbəb ola biləcək \
kodu "try" blokunda yazıb, "else" blokunda digər kodlarınızı yazmaqda davam edəsiniz.

```python
# finally
```

finally isə hər zaman işləyir. Fərqi yoxdur, "try" bloku işlədi və ya "exception" oldu. 

```python
a = 10
b = 0
try:
    result = a / b
except Exception as e:
    print(e)
else:
    print(result)
finally:
    print('I do not care. I will always run :) ')
```

finally-nin istifadə hallarından biri ilə tanış olaq. \
1. Biz hər hansı bir fayl açdıq və onun üzərində əməliyyat apardıq. \
    Sonda faylı bağlamaq üçün finally kod blokunda bunu edə bilərik. 
2. Biz faylı açdıq amma əməliyyat hər hansı bir səbəbdən uğursuz oldu. \
    Fayl bağlanmalıdır və bunu finally kod blokunda edirik.

### SQL injection 

Databazanız hack edilə və bütün gizlin məlumatlarınız ələ keçirilə bilər. \
Siz istifadəçidən input alıb və onu birbaşa sql queryə ötürsəniz problemlər yarana bilər. \
Deyək ki, siz istifadəcidən username soruşduz və useri login etmək istəyirsiniz. \
`SELECT * FROM Users WHERE UserName = sahali`

İndi təsəvvür edək ki, useriniz hackerdi və username əvəzinə sizə `sahali or 1=1` göndərdi və \
query aşağıdakı kimi oldu: \
`SELECT * FROM Users WHERE UserName = sahali or 1=1` \
Query-nin "or 1=1" həmişə true olur və sizin bütün məlumatlarınız hackerlər üçün artıq əlçatandır. \ 

Böyük ehtimalla sizi işdən qovacaqlar :) 


Bu kimi problemlərdən uzaq olmaq üçün heç vaxt inputu bir başa sql queryə vermirik, əvəzinə placeholder \
işlədirik. İnput-ları isə .execute metoduna veririk. 


```python
# Databaza yaratdıq və item url-lərin yazmağa cəhd edək. 
# kod yuxarıda yazılmış kodun birə-bir kopyasıdır
# iterate over items and extract data from each item
for item in items: # iterate over items
    tag_a = item.find('a') # find a that holds item url
    if tag_a and tag_a.has_attr('href'): # if there is a tag and it has attribute named href 
        url = f"{base_url}{tag_a['href']}" # take that and concat with base url to get valid url
        
        # write to db
        with connection.cursor() as cursor:
            cursor.execute(insert_query, (url, ))
            connection.commit()
```

```python
# copy of above code
# məlumatları dictə yığacayıq
for item in db_items[:1]:
    item_url = item['url']
    print(url)
    
    # for me
    proxy = {
        'https': 'http://192.168.0.1:3128'
    }
    # just because I am behind proxy I use this. You need run belowe code instead of this
    response = requests.get(item_url, proxies=proxy) # comment out this
    
    # for you
#     response = requests.get(request_url) #uncomment and run
    
    result_dict = dict()

    if reponse.status_code == 200:
        inner_soup = BeautifulSoup(response.content, 'lxml')
        ul_container = inner_soup.find('ul', class_='product-properties')
        li_items = ul_container.find_all('li')
        
        for li_item in li_items:
            key = li_item.find('label')
            value = li_item.find('div')
            if key: 
                key = key.text.strip()
            if value:
                value = value.text.strip()
            result_dict[key] = value
        pp(result_dict) # pretty print
        
    else:
        print('Error happened!')
```
