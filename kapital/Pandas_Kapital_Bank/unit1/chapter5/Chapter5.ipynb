## 5.4.2 The between Method

```python
higher_than_80 = employees["Salary"] >= 80000
lower_than_90 = employees["Salary"] < 90000
employees[higher_than_80 & lower_than_90].head()
```

```python
between_80k_and_90k = employees["Salary"].between(80000, 90000)
employees[between_80k_and_90k].head()
```

```python
eighties_folk = employees["Start Date"].between(
    left = "1980-01-01",
    right = "1990-01-01"
)

employees[eighties_folk].head()
```

```python
eighties_folk = employees["Start Date"].between(
    "1980-01-01",
    "1990-01-01"
)

employees[eighties_folk].head()
```

```python
filter = employees["First Name"].str[0].between('R','S')
filter
```

```python
employees.loc[filter,:]
```

```python
name_starts_with_r = employees["First Name"].between("R", "S")
employees[name_starts_with_r]['First Name'].unique()
```

```python
employees[employees["First Name"].between('R','S')]
```

## 5.4.4 Dealing with Null Values

```python
employees = pd.read_csv(
    "employees.csv", parse_dates = ["Start Date"]
)
```

```python
employees
```

```python
employees.dropna()
```

```python
employees.dropna(how = "all").tail() #butun row nulldursa silir
```

```python
employees.dropna(how = "any").tail() # rowda her hansi null varsa o rowu silir 
```

```python
employees.dropna(subset = ["Gender"]).tail() # subset olaraq genderin icinde null varsa rowu silir
```

```python
employees.dropna(subset = ["Start Date", "Salary"]).head()
```

```python
employees.dropna(how = "any", thresh = 4).head() # count(null)>=4  ise rowu silir
```
