# 10.5 Outer Joins

```python
groups.head()
```

```python
cities.head()
```

```python
groups.merge(cities,left_on="city_id",right_on="id",how="outer")
```

```python
groups.merge(cities,
             left_on="city_id",
             right_on="id",
             how="outer",
            indicator=True)
```

```python
outer_join=groups.merge(cities,
             left_on="city_id",
             right_on="id",
             how="outer",
            indicator=True)
```

```python
in_right_only=outer_join["_merge"]=="right_only"
```

```python
outer_join[in_right_only]
```
