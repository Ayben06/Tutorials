```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('q2_data.csv')
```

```python
df
```

```python
df['price']=df['price'].apply(lambda x: x.replace('$',''))
```

```python
df['price']=df['price'].apply(lambda x: x.replace(',','')).astype(float)
```

```python
df['price']
```

```python
df.groupby(['month']).agg({'order': ['max','min']})
```

```python
df.groupby(['category','colour']).agg({'price': ['sum']}).apply(lambda x: x<2000000)
```

```python
mean=df['price'].mean()
```

```python
mask=df['price'].apply(lambda x: x>mean)
```

```python
count=df.loc[mask].count()[0]
```

```python
n=df.count()[0]
```

```python
proportion=count/n
```

```python
proportion
```
